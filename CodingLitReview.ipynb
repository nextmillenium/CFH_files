```python
# Cleaning dataframe
# title, creator, subject, description, publisher,
lit_rev_grp2 = lit_rev_grp2[['Title', 'Creator', 'Subject', 'Description', 'Publisher', 'Source', 'Permalink']].copy()

lit_rev_grp2 = lit_rev_grp2.fillna("Unknown")

lit_rev_grp2
```

```python
# conduct content analysis on "description" column
# focusing on 'description' collumn by making it a series
description_grp2 = lit_rev_grp2['Description']
print(type(description_grp2))

# changing that series into a list
description_grp2 = list(lit_rev_grp2['Description'])
print(type(description_grp2))

print(description_grp2)
```

```python
# need to turn list into a string, a loop will accomplish this

description_string_grp2 = ''
for description in description_grp2:
    description_string_grp2 = description_string_grp2 + "," + description
    # the "shortened" version of the code above would be: 
    # description_string_grp2 += "," + description

# print(description_string_grp2)
# print(type(description_string_grp2))

from nltk.text import Text

description_text_grp2 = word_tokenize(description_string_grp2)
description_text_grp2 = Text(description_text_grp2)
description_text_grp2.concordance("digital" or "humanities", lines=80)








    
```
